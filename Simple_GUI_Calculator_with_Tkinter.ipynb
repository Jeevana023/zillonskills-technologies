```
import ipywidgets as widgets
from IPython.display import display

def create_calculator_gui():
    """Creates and runs the ipywidgets GUI for the calculator."""

    # Text widget to display input and results
    display_text = widgets.Text(
        value='0',
        description='',
        disabled=True,
        layout=widgets.Layout(width='98%', height='50px')
    )

    # Grid layout for the buttons
    grid = widgets.GridspecLayout(5, 4, width='99%')

    # Global variable to store the current expression
    current_expression = ""

    def button_click(b):
        """Handles number and operator button clicks."""
        nonlocal current_expression

        char = b.description

        if char == 'C':
            current_expression = ""
            display_text.value = "0"
        elif char == '=':
            try:
                # Replace 'x' with '*' for evaluation
                expression_to_eval = current_expression.replace('x', '*')
                result = str(eval(expression_to_eval))
                display_text.value = result
                current_expression = result # Allow chaining operations on result
            except Exception as e:
                display_text.value = "Error"
                current_expression = ""
        else:
            if display_text.value == "Error":
                current_expression = ""
            current_expression += str(char)
            display_text.value = current_expression

    # Define button layout
    buttons = [
        ('7', 0, 0), ('8', 0, 1), ('9', 0, 2), ('/', 0, 3),
        ('4', 1, 0), ('5', 1, 1), ('6', 1, 2), ('x', 1, 3),
        ('1', 2, 0), ('2', 2, 1), ('3', 2, 2), ('-', 2, 3),
        ('0', 3, 0), ('.', 3, 1), ('+', 3, 2), ('=', 3, 3),
        ('C', 4, 0, 1, 4) # Clear button spans 4 columns
    ]

    # Create and place buttons
    for (text, row, col, *span) in buttons:
        button = widgets.Button(
            description=text,
            layout=widgets.Layout(width='auto', height='auto')
        )
        button.on_click(button_click)
        if span:
            grid[row, col:col+span[1]] = button
        else:
            grid[row, col] = button

    # Display the calculator
    display(display_text, grid)

if __name__ == "__main__":
    create_calculator_gui()
```
